# **Import the Libraries**

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
```

# **Load the datasets**

```python
newsgroups = fetch_20newsgroups(subset='train', categories=['sci.space', 'rec.autos'])
texts = newsgroups.data[:1000]
```

# **Convert Text to Vector**

```python
vectorizer = CountVectorizer(max_features=2000, stop_words='english')
X = vectorizer.fit_transform(texts)

# Normalize for training
X = normalize(X)
X = X.toarray()  # Convert sparse matrix to dense
```

#**Build The Encoder**

```python
input_layer = Input(shape=(X.shape[1],))  # Input: 2000 words

# Encoder
encoded = Dense(512, activation='relu')(input_layer)
encoded = Dense(128, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)  # Bottleneck

# Decoder
decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(512, activation='relu')(decoded)
decoded = Dense(X.shape[1], activation='sigmoid')(decoded)

# Autoencoder model
autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

# **Train The Model**

```python
autoencoder.fit(X, X, epochs=20, batch_size=32, validation_split=0.2)
```

# **Use The Encoder to Vector Compressed Model**

```python
# Create encoder-only model
encoder = Model(input_layer, encoded)

compressed_text_vectors = encoder.predict(X)

print("Compressed representation shape:", compressed_text_vectors.shape)
```
